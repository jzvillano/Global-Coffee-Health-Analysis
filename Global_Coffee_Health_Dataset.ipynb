# End To End Pipeline

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
# Load the data
df = pd.read_csv(file_path)
```

```python
# Split the data into training and test
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Split the training data into x and y variables
x_train = train_set.drop('BMI', axis=1)
y_train = train_set['BMI']
```

```python
# Transform the data for ML computaions
from sklearn.compose import ColumnTransformer

cat_ordinal_cols = ['Sleep_Quality', 'Stress_Level']
cat_1hot_cols = ['Gender', 'Country', 'Occupation', 'Health_Issues']

x_trans_pipeline = ColumnTransformer([
    ('cat_1hot', OrdinalEncoder(handle_unknown='error'), cat_ordinal_cols),
    ('cat_ordinal', OneHotEncoder(handle_unknown='ignore'), cat_1hot_cols)

])

x_train_trans = x_trans_pipeline.fit_transform(x_train)
x_train_trans
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn import tree

# Training the Decision Tree Model
decision_model = tree.DecisionTreeRegressor()
decision_model = decision_model.fit(x_train_trans, y_train)

# Training the Linear Regression Model
linear_model = LinearRegression()
linear_model = linear_model.fit(x_train_trans, y_train)

# Training the Support Vector Regressor
from sklearn import svm
reger_model = svm.SVR()
reger_model = reger_model.fit(x_train_trans, y_train)

# Training the Neural Network Model
from sklearn.neural_network import MLPRegressor
nn_model = MLPRegressor(random_state=42, max_iter=500)
nn_model.fit(x_train_trans, y_train)

# Training the Kernel Ridge Regression Model
from sklearn.kernel_ridge import KernelRidge
krr_model = KernelRidge(alpha=1.0)
krr_model.fit(x_train_trans, y_train)
```

## Testing

```python
# Transform the test set
x_test = test_set.drop('BMI', axis=1)
y_test = test_set['BMI']

# Transform the data for ML computations
cat_ordinal_cols = ['Sleep_Quality', 'Stress_Level']
cat_1hot_cols = ['Gender', 'Country', 'Occupation', 'Health_Issues']

y_trans_pipeline = ColumnTransformer([
    ('cat_1hot', OrdinalEncoder(handle_unknown='error'), cat_ordinal_cols),
    ('cat_ordinal', OneHotEncoder(handle_unknown='ignore'), cat_1hot_cols)

])

x_test_trans = y_trans_pipeline.fit_transform(x_test)
x_test_trans
```

```python
# Make predictions on the test set
decision_model_predicitions = decision_model.predict(x_test_trans)
linear_model_predicitions = linear_model.predict(x_test_trans)
reger_model_prediction = reger_model.predict(x_test_trans)
nn_model_predictions = nn_model.predict(x_test_trans)
krr_model_predictions = krr_model.predict(x_test_trans)
```

```python
# Measure model performance
from sklearn.metrics import mean_absolute_percentage_error

decision_error = mean_absolute_percentage_error(y_test, decision_model_predicitions)
linear_error = mean_absolute_percentage_error(y_test, linear_model_predicitions)
reger_error = mean_absolute_percentage_error(y_test, reger_model_prediction)
nn_error = mean_absolute_percentage_error(y_test, nn_model_predictions)
krr_error = mean_absolute_percentage_error(y_test, krr_model_predictions)
```

```python
print(f'Decision Tree Error: {decision_error}')
print(f'Linear Regression Error: {linear_error}')
print(f"SVR Error: {reger_error}")
print(f"Neural Network Error: {nn_error}")
print(f"Kernel Ridge Regression Error: {krr_error}")
```

```python
# Create Bargraph for Mean Absolute Percentage Error

import matplotlib.pyplot as plt
import seaborn as sns

model_errors = {
    'Linear Regression': linear_error,
    'Decision Tree': decision_error,
    'SVR': reger_error,
    'Neural Network': nn_error,
    'Kernel Ridge Regression': krr_error
}

model_names = list(model_errors.keys())
error_values = list(model_errors.values())

plt.figure(figsize=(10, 6))
sns.barplot(x=model_names, y=error_values)
plt.title('Model Performance (Mean Absolute Percentage Error)')
plt.xlabel('Regression Model')
plt.ylabel('Mean Absolute Percentage Error')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
# Appeding the predictions in the data set.
test_set['BMI Predictions (Linear Model)'] = linear_model_predicitions
test_set['BMI Predictions (Decision Tree Model)'] = decision_model_predicitions
test_set['BMI Predictions (SVR Model)'] = reger_model_prediction
test_set['BMI Predictions (Neural Network Model)'] = nn_model_predictions
test_set['BMI Predictions (Kernel Ridge Regression Model)'] = krr_model_predictions
test_set
```

# End to End Pipeline for Classification

```python
#Clean Dataframe by filling null values

df['Health_Issues'] = df['Health_Issues'].fillna('No known health issues')
df['Health_Issues'].value_counts()
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size = 0.2,
                                       stratify = df['Health_Issues'], random_state= 42)
```

```python
# Split the training set into x and y variables
x_train_set = train_set.drop('Health_Issues', axis=1)
y_train_set = train_set['Health_Issues']
```

```python
x_train_set.info()
```

```python
# Split the numerical columns from the categorical columns
num_cols = x_train_set.select_dtypes(include=['float64', 'int64']).columns.tolist()
cat_ordinal_cols = ['Sleep_Quality', 'Stress_Level']
cat_1hot_cols = ['Gender', 'Country', 'Occupation']
```

```python
# Create a transformation pipeline for data processing
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

trans_pipeline_x = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ('cat_ordinal', OrdinalEncoder(handle_unknown='error'), cat_ordinal_cols),
    ('cat_1hot', OneHotEncoder(handle_unknown='ignore'), cat_1hot_cols),
])

x_train_trans = trans_pipeline_x.fit_transform(x_train_set)
```

```python
# Training the Decision Tree Model

from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
tree_model = tree_model.fit(x_train_trans, y_train_set)

# Training the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier()
forest_model = forest_model.fit(x_train_trans, y_train_set)

# Training the Ridge Classifier
from sklearn.linear_model import RidgeClassifier
ridge_model = RidgeClassifier()
ridge_model = ridge_model.fit(x_train_trans, y_train_set)

# Training the MLP Classifier or Multi-layer Perceptron
from sklearn.neural_network import MLPClassifier
mlp_model = MLPClassifier()
mlp_model = mlp_model.fit(x_train_trans, y_train_set)

# Training the Support Vector Classifier
from sklearn.svm import SVC
svc_model = SVC()
svc_model = svc_model.fit(x_train_trans, y_train_set)
```

```python
# Prepare testing set for model predictions
x_test_set = test_set.drop('Health_Issues', axis = 1)
y_test_set = test_set['Health_Issues']

# Split the numerical columns from the categorical columns
num_cols = x_test_set.select_dtypes(include=['float64', 'int64']).columns.tolist()
cat_ordinal_cols = ['Sleep_Quality', 'Stress_Level']
cat_1hot_cols = ['Gender', 'Country', 'Occupation']

# Create a transformation pipeline for data processing
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

trans_pipeline_y = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ('cat_ordinal', OrdinalEncoder(handle_unknown='error'), cat_ordinal_cols),
    ('cat_1hot', OneHotEncoder(handle_unknown='ignore'), cat_1hot_cols),
])


x_test_trans = trans_pipeline_y.fit_transform(x_test_set)
```

```python
# Model Predictions
tree_model_predictions = tree_model.predict(x_test_trans)
forest_model_predictions = forest_model.predict(x_test_trans)
ridge_model_predictions = ridge_model.predict(x_test_trans)
mlp_model_predictions = mlp_model.predict(x_test_trans)
svc_model_predictions = svc_model.predict(x_test_trans)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test_set, tree_model_predictions))
```
